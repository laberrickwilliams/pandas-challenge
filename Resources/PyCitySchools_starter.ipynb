```python
# Import Dependencies

import csv 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path


# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# PD Read School and Student Data File 
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data 
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()

#Total Number Of Unique Schools
total_schools=school_data["school_name"].count()
total_schools

#Total Students
total_students=student_data["student_name"].count()
total_students

#Total Budget
total_budget=school_data["budget"].sum()
total_budget

#Average Math Score
avg_math=student_data[("math_score")].mean()
avg_math

#Average Reading Score
avg_reading=student_data[("reading_score")].mean()
avg_reading

# % Passing math (the percentage of students who passed math)
percent_math_pass=(school_data_complete["math_score"]>70).sum() / school_data_complete["math_score"].count()* 100
percent_math_pass

# % Passing Reading (the percentage of students who passed reading)
percent_reading_pass= (school_data_complete["reading_score"]>70).sum() / school_data_complete["reading_score"].count()* 100
percent_reading_pass

# % Overall Passing (the percentage of students who passed math AND reading)
overall_avg=((avg_math + avg_reading) /2)
overall_avg

#School Name
school_count= len(school_data_complete["school_name"].unique())
school_count

#School Type
school_types=school_data.set_index(["school_name"])["type"]

#Total Students
student_per_school=school_data_complete["school_name"].value_counts()
student_per_school.head()

#Total School Budget
school_budget=school_data_complete.groupby(["school_name"])["budget"].mean()
school_budget.head()

#Per Student Budget
per_student_budget=school_budget/student_per_school
per_student_budget

#Average Math Score
avg_math_perSchool=school_data_complete.groupby(["school_name"])["math_score"].mean()
avg_math_perSchool.head()

#Average Reading Score
avg_reading_perSchool= school_data_complete.groupby(["school_name"])["reading_score"].mean()
avg_reading_perSchool.head()

# % Passing Math (the percentage of students who passed math)
pass_math_score=school_data_complete.loc[school_data_complete["math_score"]>=70]
pass_math_score
group_math_school=pass_math_score["school_name"].value_counts()
group_math_school.head()
percent_math=group_math_school/student_per_school*100
percent_math

# % Passing Reading (the percentage of students who passed reading)
read_scor=school_data_complete.loc[school_data_complete["reading_score"]>=70]
by_read_PerSch= read_scor["school_name"].value_counts()
by_read_PerSch.head()
percent_read=by_read_PerSch/stu_per_sch*100
percent_read

# % Overall Passing (the percentage of students who passed math AND reading)
overall=percent_math + percent_read/stu_per_sch
overall

#School Summary
school_summery_df = pd.DataFrame({"School Type":school_types, "Total Students":stu_per_sch, "Total School Budget":school_budget,"Per Student Budget":per_student_budget,"Average Math Score" :avg_math_perSch,"Average Reading Score":avg_reading_perSchool,"% Passing Math":percent_math, "% Passing Reading":percent_read,  "% Overall Passing Rate":overall})
school_summery_df.head()

#Highest-Performing Schools (by % Overall Passing)
TopSch_df=school_summery_df.sort_values(["% Passing Math","% Passing Reading","% Overall Passing Rate" ], ascending=False)
TopSch_df.head()

#Lowest-Performing Schools (by % Overall Passing)
BottomSch_df=school_summery_df.sort_values("% Overall Passing Rate")
BottomSch_df.head()

#Math Scores by Grade
math_9th=student_data.loc[student_data["grade"] == "9th"].groupby(["school_name"])['math_score'].mean()
math_9th.head()
math_10th=student_data.loc[student_data["grade"] == "10th"].groupby(["school_name"])['math_score'].mean()
math_10th.head()
math_11th=student_data.loc[student_data["grade"] == "11th"].groupby(["school_name"])["math_score"].mean()
math_11th.head()
math_12th=student_data.loc[student_data["grade"] == "12th"].groupby(["school_name"])["math_score"].mean()
math_12th.head()
#Math Scores DF
math_byGrad=pd.DataFrame({"9th":math_9th, "10th": math_10th, "11th":math_11th, "12th":math_12th })
math_byGrad

#Reading Scores by Grade
read_9th=student_data.loc[student_data["grade"] == "9th" ].groupby(["school_name"])["reading_score"].mean()
read_10th=student_data.loc[student_data["grade"] == "10th" ].groupby(["school_name"])["reading_score"].mean()
read_11th=student_data.loc[student_data["grade"] == "11th" ].groupby(["school_name"])["reading_score"].mean()
read_12th=student_data.loc[student_data["grade"] == "12th" ].groupby(["school_name"])["reading_score"].mean()
#Reading Scores DF
reading_byGrade=pd.DataFrame({"9th":read_9th,"10th":read_10th, "11th":read_11th, "12th":read_12th })
reading_byGrade

#Scores by School Spending
school_summery_df.head()
#Bins
bins = [0, 585, 630, 645, 680]
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]
school_summery_df["Spending Ranges (Per Student)"]=pd.cut(school_summery_df["Total School Budget"]/school_summery_df["Total Students"],bins, labels=group_names)
school_summery_df.head()
scores_by_sch=school_summery_df.drop(columns=["Total Students", "Total School Budget","School Type", "Per Student Budget"])
scores_by_sch

pd.cut

#mean per Math Scores
by_spending =scores_by_sch.groupby(["Spending Ranges (Per Student)"])
by_spending.mean()

#School Summary Header
school_summery_df.head()

#Scores by School Size bins & lables 
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
type("School Size")
scores_by_sch["School Size"]=pd.cut(school_summery_df["Total Students"],size_bins , labels=group_names)
scores_by_sch.head()

#Scores by School Type
school_summery_df.head()



```
